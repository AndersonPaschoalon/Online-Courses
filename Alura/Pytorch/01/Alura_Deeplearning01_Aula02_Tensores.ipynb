```python
# Tensor A partir de listas

import torch


ll = [[1, 2, 3], [4, 5, 6]]

ll = [[1, 2, 3], [4, 5, 6]]
tns = torch.Tensor(ll)
print(tns.dtype)
print(tns)

tns = torch.DoubleTensor(ll)
print(tns.dtype)
print(tns)

tns = torch.LongTensor(ll)
print(tns.dtype)
print(tns)
```
