```python
import numpy as np
import matplotlib.pyplot as plt
# equação da reta

a = -1
b = 4
c = 0.4

def plot_line(a, b, c):
  # espaço amostral
  x = np.linspace(-2, 4, 50)

  # ax + by + c = 0
  # y = (-a*x -c)/b
  y = (-a*x - c)/b

  print(x)
  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2, 4, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)

plot_line(a, b, c)

```

```python
np.random.seed(46)
x, y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
)
plt.scatter(x[:,0], x[:,1], marker='o', c=y, edgecolor='k')

ymin, ymax = plt.gca().get_ylim()
xmin, xmax = plt.gca().get_xlim()

def plot_model(w1, w2, b):
  # espaço amostral
  x = np.linspace(-2, 4, 50)

  y = (-w1*x - b)/w2

  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2, 4, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)

plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

w1 = 5
w2 = 1
b = -0.4

plot_model(w1, w2, b)

p = (-1, 1)



def classify(ponto, w1, w2, b):
  ret = w1 * ponto[0] + w2* ponto[1] + b

  if ret >= 0:
    return 1, 'yellow'
  else:
    return 0, 'blue'


p = (2, -1)

classe, cor = classify(p, w1, w2, b)
print(classe, cor)

plot_model(w1, w2, b)

plt.plot(p[0], p[1], marker='^', color=cor, markersize=20)

acertos = 0
for k in range(len(x)):
  cat, cor = classify(x[k], w1, w2, b)
  if cat == y[k]:
    acertos += 1

print("Acuracia: {0}%".format(100*acertos/len(x)))


```
