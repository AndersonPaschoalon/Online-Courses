```python
import torch 
from torch import nn

if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')


print(device)
```

```python
from sklearn import datasets

wine = datasets.load_wine()
data = wine.data
target = wine.target

print(data.shape, target.shape)
print(wine.feature_names, wine.target_names, target.shape)
```

```python

class WineClassifier(nn.Module):

  def __init__(self, input_size: int, hidden_size: int, out_size: int):
    # inicialização obrigatoria da classe base
    super(WineClassifier, self).__init__()
    # definição da arquitetura dessa modelagem
    self.hidden = nn.Linear(input_size, hidden_size)
    self.relu = nn.ReLU()
    self.out = nn.Linear(hidden_size, out_size)
    self.softmax = nn.Softmax()

  def forward(self, X):
    feature = self.relu(self.hidden(X))
    output = self.softmax(self.out(feature))
    return output

# definir device
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
print(device)

# criar rede neural
input_size = data.shape[1]
hidden_size = 32
out_size = len(wine.target_names)
print("input_size:" + str(input_size) + ", hidden_size:" + str(hidden_size) + ", out_size:" + str(out_size) + " => " + str(device))
net = WineClassifier(input_size, hidden_size, out_size).to(device)
print(net)


```

```python
criterion = nn.CrossEntropyLoss().to(device) # cast na GPU
Xtensor = torch.from_numpy(data).float()
Ytensor = torch.from_numpy(target)

# Cast na GPU
Xtensor = Xtensor.to(device)
Ytensor = Ytensor.to(device)

print(Xtensor.dtype, Ytensor.dtype)
```

```python
## pasando uma unica amostra
#pred = net(Xtensor[0].unsqueeze(0))
#print(Xtensor[0].shape)
#print(pred.shape)

# passando todas as amostras
pred = net(Xtensor)
print(Xtensor.shape)
print(pred.shape)
print(pred[0].data, Ytensor[0].data)

print("Função de perda para um dado:")
loss = criterion(pred[0].unsqueeze(0), Ytensor[0].unsqueeze(0))
print("=> loss:" + str(loss))

print("Função de perda para todos os dados:")
loss = criterion(pred[:30], Ytensor[:30])
print("=> loss:" + str(loss))


```

```python
from sklearn import datasets

diabetes = datasets.load_diabetes()
data = diabetes.data
target = diabetes.target

print(data.shape, target.shape)

print(data[0]) 
print(target[0])
```

```python
import torch 
from torch import nn

class WineClassifier(nn.Module):

  def __init__(self, input_size: int, hidden_size: int, out_size: int):
    # inicialização obrigatoria da classe base
    super(WineClassifier, self).__init__()
    # definição da arquitetura dessa modelagem
    self.hidden = nn.Linear(input_size, hidden_size)
    self.relu = nn.ReLU()
    self.out = nn.Linear(hidden_size, out_size)
    self.softmax = nn.Softmax()

  def forward(self, X):
    feature = self.relu(self.hidden(X))
    output = self.softmax(self.out(feature))
    return output

# definir device
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
print(device)

# criar rede neural
print("# criar rede neural")
input_size = data.shape[1]
hidden_size = 32
out_size = 1
print("input_size:" + str(input_size) + ", hidden_size:" + str(hidden_size) + ", out_size:" + str(out_size) + " => " + str(device))
net = WineClassifier(input_size, hidden_size, out_size).to(device)
print(net)

##############

criterion = nn.MSELoss().to(device)

# cast das variaveis
print("# cast das variaveis")
Xtensor = torch.from_numpy(data).float().to(device)
Ytensor = torch.from_numpy(target).float().to(device)

print(Xtensor.shape, Ytensor.shape)

# predição

print("# predição")
pred = net(Xtensor)
#print(pred.shape)
loss = criterion(pred.squeeze(), Ytensor)
print("loss:" + str(loss))
```
